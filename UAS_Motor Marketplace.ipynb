```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import library pandas, numpy, dan matplotlib

```python
data = pd.read_csv("motor.csv")
data
```

Membaca file dataset

```python
data.info()
```

Memberi kejelasan dataset dari aspek kolom dan tipe data

```python
from sklearn.preprocessing import LabelEncoder
for col in ['price','mileage','power']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
data
```

Mengubah tipe data float menjadi integer

```python
data = data[["price","mileage"]]
data.head(10)
```

Mengambil 10 data teratas

```python
data.describe()
```

Menampilkan count, mean, std, min dalam data price dan mileage

```python
fig, ax = plt.subplots()
ax.boxplot(data,
           vert = False,
           showmeans = True,
           meanline = True,
           labels =("price","mileage"),
           patch_artist = True,
           medianprops = {"linewidth" : 2, "color" : "red"},
            meanprops = {"linewidth" : 2, "color" : "yellow"})
plt.show()
```

Membuat visual outside

```python
kolom = ["price","mileage"]
Q1 = data[kolom].quantile(0.25)
Q3 = data[kolom].quantile(0.75)
IQR = Q3-Q1
data = data[~((data[kolom]<(Q1 - 1.5*IQR)) |
              (data[kolom]>(Q3 + 1.5*IQR))).any(axis=1)]

plt.boxplot(data[kolom])
plt.xticks([1,2],kolom)
plt.title("Sesudah Outline Terbuang")
plt.show()
```

Menampilkan outside yang sudah dibuang

```python
data.info()
```

Mendeskripsikan tipe data dari price dan mileage

```python
x_array = np.array(data)
```

Cara mengubah data yang berasal dari pandas DataFrame atau struktur data lainnya menjadi array NumPy

```python
plt.scatter(data.price, data.mileage, c='magenta')
plt.show()
```

Menampilkan grafik dari data price dan mileage

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_array)
```

```python
from sklearn.cluster import KMeans
sse = []
index = range(1,10)
for i in index :
    kmeans = KMeans(n_clusters = i, random_state = 0, n_init = 'auto')
    kmeans.fit(x_scaled)
    sse_ = kmeans.inertia_
    sse.append(sse_)
    print (i,sse_)
```

```python
kmeans = KMeans (n_clusters = 3, random_state = 0)
kmeans.fit (x_scaled)
```

Menggunakan algoritma k-means clustering dari scikit-learn (sklearn) untuk melakukan klastering pada data yang telah di-scaling. Setelah menjalankan kedua perintah ini, model k-means (kmeans) akan terlatih dan dapat digunakan untuk melakukan prediksi atau mengevaluasi klaster yang dihasilkan dari data yang telah di-scaling (x_scaled)

```python
kmeans.cluster_centers_
```

Penentuan nilai tengah dengan array

```python
output = plt.scatter(x_scaled[:,0],x_scaled[:,1], s=10, c = kmeans.labels_)
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="yellow", s=10)
plt.title("Motor Marketplace KMeans Clustering")
plt.xlabel("Price")
plt.ylabel("Mileage")
plt.colorbar(output)
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
labels = kmeans.labels_
davies_bouldin_score(x_scaled, labels)
```

Score dari clustering KMeans, skala ini juga memberikan ukuran seberapa baik suatu clustering bekerja berdasarkan rata-rata dari rasio kesamaan antara setiap klaster dengan klaster lainnya
